# Seaborn :- 

Seaborn is a well-known Python library for data visualization that offers a user-friendly interface for producing visually appealing and informative statistical graphics. It is designed to work with Pandas DataFrames, making it easy to visualize and explore data quickly and effectively.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt 
```

# Loading a Dataset :-


Seaborn provides built-in datasets that can be easily loaded for practice, demonstration, and educational purposes.


```python
data = sns.load_dataset("iris")
data.head(10)
```

# Line Plot :- 

A line plot is used to display data points connected by straight lines.

```python
data = {'Year': [2015, 2016, 2017, 2018, 2019], 'Sales': [120, 150, 170, 160, 180]}
df = pd.DataFrame(data)

sns.lineplot(x='Year', y='Sales', data=df)
plt.show()
```

```python
data = sns.load_dataset("flights")
sns.lineplot(x="year", y="passengers", data=data)
plt.show()
```

# Histogram :- 

A histogram is a classic visualization tool that represents the distribution of one or more variables by counting the number of observations that fall within discrete bins.

```python
penguins = sns.load_dataset("penguins")
sns.histplot(data=penguins, x="bill_length_mm", bins=20, kde=True)
plt.show()
```

# Bar Plot :- 

A bar plot represents an aggregate or statistical estimate for a numeric variable with the height of each rectangle and indicates the uncertainty around that estimate using an error bar.

```python
titanic = sns.load_dataset("titanic")
sns.barplot(x="class", y="survived", data=titanic)
plt.show()
```

# Heatmap :- 

A heatmap uses colored cells to show the correlation between two discrete dimensions, which is very important in feature selection.

```python
df = sns.load_dataset("iris")
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", linewidths=0.5)
plt.show()
```

# Scatter Plot:- 

A scatter plot is used to display values for two variables for a set of data.

```python
mpg = sns.load_dataset("mpg")
sns.scatterplot(data=mpg, x="horsepower", y="mpg", hue="origin", style="origin")
plt.show()
```

# KDE Plot:- 

A KDE (Kernel Density Estimation) plot is a statistical tool that generates a smoothed, continuous probability density function (PDF) based on a set of data points.

```python
data = sns.load_dataset("iris")
sns.kdeplot(x="sepal_width", data=data, hue="species", fill=True, common_norm=False, alpha=0.5)
plt.show()
```

# Rug Plot :- 

The Seaborn rugplot() method is used to plot marginal distributions and allows drawing ticks along the x and y axes.

```python
data = sns.load_dataset("iris")
sns.kdeplot(data=data, x="petal_length")
sns.rugplot(data=data, x="petal_length", height=0.1)
plt.show()
```

# Box Plot :- 

A box plot shows the distribution of quantitative data in a way that facilitates comparisons between variables or across levels of a categorical variable.

```python
titanic = sns.load_dataset("titanic")
sns.boxplot(x="class", y="age", data=titanic, hue="sex")
plt.show()
```

# Violin Plot :- 

A violin plot combines aspects of a box plot and a density plot to show the distribution of the data.

```python
tips = sns.load_dataset("tips")
sns.violinplot(x="day", y="total_bill", data=tips, hue="time", split=True)
plt.show()
```

# Pair Plot :- 

A pair plot is used to visualize pairwise relationships in a dataset.

```python
data = sns.load_dataset("penguins")
sns.pairplot(data, hue="species")
plt.show()
```
