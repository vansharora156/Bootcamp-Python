# NUMPY :-

1. Create a 5x5 array with random values and normalize it

```python
import numpy as np
array = np.random.random((5, 5))
normalized_array = (array - np.min(array)) / (np.max(array) - np.min(array))
print(normalized_array)
```

2. Given an array, replace all elements greater than a given value with that value

```python
import numpy as np
array = np.array([1, 5, 10, 15, 20])
value = 10
array[array > value] = value
print(array)
```

3. Create a 2D array and calculate the sum of each row

```python
import numpy as np
array = np.random.randint(0, 10, (3, 4))
row_sums = np.sum(array, axis=1)
print(row_sums)
```

4. Generate a 4x4 matrix and flatten it to a 1D array

```python
import numpy as np
matrix = np.random.randint(0, 10, (4, 4))
flat_array = matrix.flatten()
print(flat_array)
```

5. Find the unique elements in an array

```python
import numpy as np
array = np.array([1, 2, 2, 3, 4, 4, 5])
unique_elements = np.unique(array)
print(unique_elements)
```

# PANDAS :-

1. Read a CSV file and display the first 5 rows

```python
import pandas as pd
df = pd.read_csv('file.csv')
print(df.head())
```

2. Filter rows where a specific column has values greater than a given threshold 

```python
import pandas as pd
df = pd.DataFrame({'A': [1, 2, 3, 4, 5], 'B': [5, 4, 3, 2, 1]})
filtered_df = df[df['A'] > 3]
print(filtered_df)
```

3. Add a new column to a DataFrame based on the values of an existing column

```python
import pandas as pd
df = pd.DataFrame({'A': [1, 2, 3, 4, 5]})
df['B'] = df['A'] * 2
print(df)
```

4. Group a DataFrame by a column and calculate the mean of each group 

```python
import pandas as pd
df = pd.DataFrame({'A': ['foo', 'bar', 'foo', 'bar'], 'B': [1, 2, 3, 4]})
group_mean = df.groupby('A').mean()
print(group_mean)
```

5. Handle missing data by filling with the mean value of the column

```python
import pandas as pd
df = pd.DataFrame({'A': [1, 2, None, 4], 'B': [None, 2, 3, 4]})
df['A'].fillna(df['A'].mean(), inplace=True)
df['B'].fillna(df['B'].mean(), inplace=True)
print(df)
```

# Matplotib :-

1. Create a line plot with Matplotib

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
plt.plot(x, y)
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

2. Create a bar plot 

```python
import matplotlib.pyplot as plt
categories = ['A', 'B', 'C']
values = [4, 7, 1]
plt.bar(categories, values)
plt.title("Bar Plot")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

3. Create a scatter plot with Matplotlib

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
plt.scatter(x, y)
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

4. Create a subplot with 2 rows and 1 column in Matplotlib

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
y2 = [1, 4, 6, 8, 10]

plt.subplot(2, 1, 1)
plt.plot(x, y1)
plt.title("First Plot")

plt.subplot(2, 1, 2)
plt.plot(x, y2)
plt.title("Second Plot")

plt.tight_layout()
plt.show()
```

5. Customize the line style and color in a Matplotlib plot

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
plt.plot(x, y, linestyle='--', color='r')
plt.title("Customized Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

# Seaborn :-

1. Create a scatter plot 

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")
sns.scatterplot(data=tips, x="total_bill", y="tip")
plt.title("Scatter Plot")
plt.show()
```

2. Create a heatmap 

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
data = np.random.rand(10, 12)
sns.heatmap(data, annot=True, fmt=".1f")
plt.title("Heatmap")
plt.show()
```

3. Create a box plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")
sns.boxplot(data=tips, x="day", y="total_bill")
plt.title("Box Plot")
plt.show()
```

4. Change the color palette

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_palette("pastel")
tips = sns.load_dataset("tips")
sns.boxplot(data=tips, x="day", y="total_bill")
plt.title("Box Plot with Custom Palette")
plt.show()
```

5. Use the hue parameter in Seaborn to add a third dimension to a plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time")
plt.title("Scatter Plot with Hue")
plt.show()
```
