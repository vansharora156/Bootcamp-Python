# Handling Null Values :-

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv(r"C:\Users\DELL\Downloads\titanic_train.csv")
df.head(10)
```

isnull() :- 

This method returns a DataFrame of the same shape as the input, where each element is True if it's a null value and False otherwise. It's useful for pinpointing null values within the dataset.

```python
df.isnull().sum()  #gives sum of null values in a specific column
```

info() :- 

The info() method provides a concise summary of the DataFrame, including the count of non-null values for each column. It's a quick way to assess the presence of null values and the overall data structure.

```python
df.info()
```

fillna(): Fills null values with a specified value or method.

```python
df2 = df["Embarked"].fillna(value = "S") 
df2.head()

df2.isnull().sum()
# Check the number of null values in the entire DataFrame df2
print(df2.isnull().sum().sum())

```

ONE OF THE WAY TO REMOVE NULL VALUES ARE USING PAD WHICH MEANS FILLING IT WITH PREVIOUS ROWS DATA,BFIll and many more down

```python
df3 = df.fillna(method='pad')
df3 #the output will have the Nan value replaced with the value of the row above it.
```

```python
# Let's change the null value with the value of the next row
df4 = df.fillna(method='bfill')
df4 
```

```python
# Filling null values with the previous value but column wise
df5 = df.fillna(method='pad',axis=1)
df5
```

```python
df6 = df.fillna({ 'Cabin': 'Unknown'})
print(df6.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(16, 8)) 

# Histogram
plt.subplot(1, 2,1 )
sns.histplot(df['Fare'], kde=True)
plt.title('Histogram of Fare')
plt.xlim(0, 100)

# Density Plot (KDE)
plt.subplot(1, 2, 2)
sns.kdeplot(df['Fare'], shade=True)
plt.title('Density Plot of Fare')
plt.xlim(0, 100)

plt.tight_layout()
plt.show()
```

```python
median_fare = df['Fare'].median()
print("Median Fare:", median_fare)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
df['Fare_filled'] = df['Fare'].fillna(df['Fare'].median())
fig, axs = plt.subplots(1, 2, figsize=(16, 6), sharey=True)  

sns.histplot(df['Fare'], kde=True, color='lightblue', bins=40, ax=axs[0])
axs[0].set_title("Original Data Distribution")
axs[0].set_xlabel("Fare")
axs[0].set_ylabel("Frequency")
axs[0].set_xlim(0, 100)  

sns.histplot(df['Fare_filled'], kde=True, color='green', bins=40, ax=axs[1])
axs[1].set_title("Data Distribution after Filling with Median")
axs[1].set_xlabel("Fare")
axs[1].set_xlim(0, 100) 

plt.tight_layout()
plt.show()
```

dropna():

The dropna() method in pandas removes rows or columns with null values from a DataFrame. Parameters like subset, thresh, how, and axis offer customization:

```python
df10 = df.dropna()
df10
```

```python
# how='any' means if any row has a null values it is removed (default)
df11 =df.dropna(how='any')
df11
```

```python
# how='all' means it drop the row that have all nul values
df12 = df.dropna(how='all')
df12
```

```python
# What if you want to drop column with null values, there is a parameter for it
df13 = df.dropna(axis=1)
df13
```
